# Getting Data

This is a preliminary notebook to explore the Pandas Datareader functionality to get options data, and explore the datastructure that it provides.

```python
from pandas_datareader import Options

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.set_context('paper')

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (5,5/1.618)
```

```python
aapl = Options('aapl', 'yahoo')
```

```python
aapl.expiry_dates
```

```python
data = aapl.get_all_data()
```

```python
feb_options = data.loc[(slice(None), '2018-04-20', slice(None)),:]
```

```python
feb_options[feb_options.Vol > 100]
```

```python
feb_calls = feb_options.loc[(slice(None), slice(None), 'call'), :]
feb_puts = feb_options.loc[(slice(None), slice(None), 'put'), :]

plt.plot(feb_calls.index.get_level_values('Strike'),
    feb_calls.loc[(slice(None), slice(None), 'call'), :].Vol,
        label='Calls')
plt.plot(feb_puts.index.get_level_values('Strike'),
    feb_puts.loc[(slice(None), slice(None), 'put'), :].Vol,
        label='Puts')
plt.title('Volume Traded')
plt.xlim((100,220))
plt.ylim((0,40000))
plt.axvline(aapl.underlying_price, c='k', label='Spot')
plt.legend()
```

```python
plt.plot(feb_calls.index.get_level_values('Strike'),
    feb_calls.loc[(slice(None), slice(None), 'call'), :].IV,
        label='Calls')
plt.plot(feb_puts.index.get_level_values('Strike'),
    feb_puts.loc[(slice(None), slice(None), 'put'), :].IV,
        label='Puts')
plt.title('AAPL Apr20 Implied Volatility')
plt.ylabel('Implied Volatility')
plt.xlabel('Strike Price')
plt.xlim((100,220))
plt.ylim((0,.8))
plt.axvline(aapl.underlying_price, c='k', label='Spot', linewidth=1)
sns.despine()
plt.legend()
plt.tight_layout()
plt.savefig('report/figs/aapliv.eps')
```
